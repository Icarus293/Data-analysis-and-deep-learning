## Bài 4
Viết chương trình Python với thư viện NLTK để loại bỏ các stopword từ một văn bản đã cho.


```python
text = (
    "This is a simple example to demonstrate how to remove stopwords "
    "from a given text using the NLTK library."
)

sw_en = set(stopwords.words('english'))
tokens = nltk.word_tokenize(text)
filtered_tokens = [w for w in tokens if w.lower() not in sw_en]

print("Văn bản gốc:")
print(tokens)
print("\nVăn bản sau khi loại bỏ stopword:")
print(filtered_tokens)
print("\nCâu sau khi loại bỏ stopword:")
print(" ".join(filtered_tokens))
```

## Bài 5
Viết chương trình Python với thư viện NLTK bỏ qua (loại trừ) một số từ ra khỏi danh sách các stopword
(ví dụ: giữ lại các từ phủ định vì chúng có thể mang ý nghĩa quan trọng khi phân tích cảm xúc).


```python
# Danh sách stopword gốc
sw_full = set(stopwords.words('english'))

# Các từ muốn GIỮ LẠI (không coi là stopword) - ví dụ từ phủ định
words_to_keep = {'not', 'no', 'never', 'nor', "n't"}

# Danh sách stopword mới đã bỏ qua (loại trừ) các từ trên
sw_custom = sw_full - words_to_keep

print(f"Số stopword gốc   : {len(sw_full)}")
print(f"Số stopword tùy chỉnh (đã bỏ qua {len(words_to_keep)} từ): {len(sw_custom)}")

text2 = "I am not happy with this product, it is no good and never works properly."
tokens2 = nltk.word_tokenize(text2)

filtered_full = [w for w in tokens2 if w.lower() not in sw_full]
filtered_custom = [w for w in tokens2 if w.lower() not in sw_custom]

print("\nLọc với stopword gốc      :", filtered_full)
print("Lọc với stopword tùy chỉnh:", filtered_custom)
```

## Bài 6
Viết chương trình Python với thư viện NLTK để tìm định nghĩa và ví dụ của một từ đã cho bằng WordNet.


```python
from nltk.corpus import wordnet as wn

word = "happy"
synsets = wn.synsets(word)

print(f"Từ '{word}' có {len(synsets)} synset (nghĩa):\n")
for s in synsets:
    print(f"- {s.name()}")
    print(f"    Định nghĩa: {s.definition()}")
    print(f"    Ví dụ     : {s.examples()}")
```

## Bài 7
Viết chương trình Python với thư viện NLTK để tìm tập hợp các từ đồng nghĩa và trái nghĩa của một từ nào đó.


```python
def get_synonyms_antonyms(word):
    synonyms = set()
    antonyms = set()
    for syn in wn.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name().replace('_', ' '))
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name().replace('_', ' '))
    return synonyms, antonyms

target_word = "good"
syn_set, ant_set = get_synonyms_antonyms(target_word)

print(f"Từ đồng nghĩa của '{target_word}': {syn_set}")
print(f"Từ trái nghĩa của '{target_word}': {ant_set}")
```

```python
noun1 = wn.synset('cat.n.01')
noun2 = wn.synset('dog.n.01')

print(f"So sánh danh từ '{noun1.name()}' và '{noun2.name()}':")
print(f"- Path similarity     : {noun1.path_similarity(noun2):.4f}")
print(f"- Wu-Palmer similarity: {noun1.wup_similarity(noun2):.4f}")
print(f"- Lch similarity      : {noun1.lch_similarity(noun2):.4f}")
```

## Bài 11
Viết chương trình Python với thư viện NLTK để tìm số lượng tên nam và nữ trong kho ngữ liệu `names`
(male.txt ~2943 tên, female.txt ~5001 tên - biên soạn bởi Kantrowitz, Ross). In ra 10 tên nam và 10 tên nữ đầu tiên.


```python
from nltk.corpus import names

male_names = names.words('male.txt')
female_names = names.words('female.txt')

print(f"Số lượng tên nam : {len(male_names)}")
print(f"Số lượng tên nữ  : {len(female_names)}")
print(f"Tổng số tên      : {len(male_names) + len(female_names)}")

print("\n10 tên nam đầu tiên:", male_names[:10])
print("10 tên nữ đầu tiên :", female_names[:10])
```

## Bài 12
Viết chương trình Python với thư viện NLTK để in 15 kết hợp ngẫu nhiên đầu tiên
được gắn nhãn 'male' và 'female' từ kho tên.


```python
labeled_names = (
    [(name, 'male') for name in male_names] +
    [(name, 'female') for name in female_names]
)

random.seed(42)  # để kết quả có thể tái lập
random.shuffle(labeled_names)

print("15 tên đã gắn nhãn ngẫu nhiên đầu tiên:")
for name, gender in labeled_names[:15]:
    print(f"  ({name!r}, {gender!r})")
```

## Bài 13
Viết chương trình Python với thư viện NLTK để trích xuất ký tự cuối cùng của tất cả các tên đã gắn nhãn
và tạo mảng mới với chữ cái cuối cùng của mỗi tên cùng với nhãn (giới tính) được liên kết.


```python
# Mảng mới: (ký tự cuối cùng của tên, nhãn giới tính)
last_letter_pairs = [(name[-1].lower(), gender) for (name, gender) in labeled_names]

print("15 phần tử đầu tiên của mảng (ký tự cuối, nhãn):")
print(last_letter_pairs[:15])

# Dạng feature-dict thường dùng để huấn luyện bộ phân loại giới tính theo ký tự cuối
def gender_features(word):
    return {'last_letter': word[-1].lower()}

featuresets = [(gender_features(name), gender) for (name, gender) in labeled_names]
print("\n5 featureset đầu tiên (dùng cho bài toán phân loại giới tính):")
print(featuresets[:5])
```
