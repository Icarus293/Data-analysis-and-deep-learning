###Sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần

```python
df_sorted_DH1 = df.sort_values(by="DH1",ascending=True)
df_sorted_DH1.head()
```

### Sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính

```python
df_sorted_DH2_by_GT = df.groupby('GT').apply(lambda x: x.sort_values(by='DH2', ascending=True))
display(df_sorted_DH2_by_GT.head())
```

### Thống kê DH1 theo KT

```python
pivot_table_KT = df.pivot_table(
    values='DH1',
    index='KT',
    aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std',
             lambda x: x.quantile(0.25),
             lambda x: x.quantile(0.5),
             lambda x: x.quantile(0.75)]
)
pivot_table_KT.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q2', 'Q3']
display(pivot_table_KT)
```

### Thống kê DH1 theo KT và KV

```python
pivot_table_KT_KV = df.pivot_table(
    values='DH1',
    index=['KT', 'KV'],
    aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std',
             lambda x: x.quantile(0.25),
             lambda x: x.quantile(0.5),
             lambda x: x.quantile(0.75)]
)
pivot_table_KT_KV.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q2', 'Q3']
display(pivot_table_KT_KV)
```

### Thống kê DH1 theo KT, KV và DT

```python
pivot_table_KT_KV_DT = df.pivot_table(
    values='DH1',
    index=['KT', 'KV', 'DT'],
    aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std',
             lambda x: x.quantile(0.25),
             lambda x: x.quantile(0.5),
             lambda x: x.quantile(0.75)]
)
pivot_table_KT_KV_DT.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q2', 'Q3']
display(pivot_table_KT_KV_DT)
```

### 1. Vẽ biểu đồ đường Simple cho biến T1

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a simple line plot for T1
plt.figure(figsize=(12, 6))
sns.lineplot(x=df.index, y='T1', data=df)
plt.title('Biểu đồ đường Simple của biến T1')
plt.xlabel('Chỉ mục')
plt.ylabel('Điểm T1')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### 2. Tạo biến phân loại `phanlopt1` cho môn toán (T1)

```python
def classify_T1(score):
    if 0 <= score < 5:
        return 'k'
    elif 5 <= score < 7:
        return 'tb'
    elif 7 <= score < 8:
        return 'k'
    elif score >= 8:
        return 'g'
    else:
        return None # Handle cases outside the defined range if any

df['phanlopt1'] = df['T1'].apply(classify_T1)
display(df[['T1', 'phanlopt1']].head())
```

### 3. Lập bảng tần số cho biến `phanlopt1`

```python
# Lập bảng tần số cho biến phanlopt1
freq_table_phanlopt1 = df['phanlopt1'].value_counts().reset_index()
freq_table_phanlopt1.columns = ['Phân loại T1', 'Tần số']
freq_table_phanlopt1['Tần suất (%)'] = (df['phanlopt1'].value_counts(normalize=True) * 100).reset_index(drop=True)

display(freq_table_phanlopt1)
```

### 4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến `phanlopt1`

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a multiple line plot for T1, classified by phanlopt1
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x=df.index, y='T1', hue='phanlopt1', palette='deep')
plt.title('Biểu đồ đường Multiple Line của T1 theo phân loại phanlopt1')
plt.xlabel('Chỉ mục')
plt.ylabel('Điểm T1')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Phân loại T1')
plt.show()
```

### 5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến `phanlopt1`

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))

# Create a stripplot to show individual T1 scores for each phanlopt1 category
sns.stripplot(data=df, x='phanlopt1', y='T1', hue='phanlopt1', jitter=True, palette='deep', dodge=True)

plt.title('Biểu đồ Stripplot của T1 theo phân loại phanlopt1')
plt.xlabel('Phân loại T1')
plt.ylabel('Điểm T1')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Phân loại T1')
plt.show()
```
