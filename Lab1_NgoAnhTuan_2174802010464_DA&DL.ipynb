### 2. Định nghĩa các thang đo phù hợp cho từng biến số

Dựa trên phân loại dữ liệu, chúng ta có thể định nghĩa các thang đo như sau:

**Thang đo định danh (Nominal Scale):** Dùng để phân loại mà không có thứ tự hoặc khoảng cách giữa các giá trị.

*   **`STT`**: Chỉ là mã định danh.
*   **`HỌ TÊN`**: Tên riêng.
*   **`GIOI TINH`**: Nam/Nữ.
*   **`DT`**: Các nhóm dân tộc khác nhau.
*   **`KV`**: Các khu vực khác nhau (ví dụ: KV1, KV2, KV3).

**Thang đo thứ bậc (Ordinal Scale):** Dùng để phân loại có thứ tự, nhưng khoảng cách giữa các giá trị không đồng đều hoặc không có ý nghĩa.

*   **`XL1`, `XL2`, `XL3`**: Xếp loại học lực (ví dụ: Kém < Yếu < Trung bình < Khá < Giỏi).

*   **`NGÀY SINH`**: Dữ liệu ngày tháng. Có thể tính khoảng cách giữa các ngày nhưng không có điểm 0 tuyệt đối.

**Thang đo tỉ lệ (Ratio Scale):** Dùng cho dữ liệu có thứ tự, khoảng cách đều và có điểm 0 tuyệt đối, cho phép so sánh tỷ lệ.

*   **`TBM1`, `TBM2`, `TBM3`**: Điểm trung bình môn học.
*   **`CN`**: Điểm tổng kết chung.
*   **`KT`**: Kết quả thi.
*   **`DH1`, `DH2`, `DH3`**: Điểm thi đại học.
*   **`KQ`**: Kết quả cuối cùng.

### 4. Thống kê dữ liệu thiếu cho cột `DT` (Dân tộc) và hiệu chỉnh dữ liệu thiếu

Chúng ta sẽ kiểm tra các giá trị thiếu trong cột `DT` và thay thế chúng bằng 0 như yêu cầu.

```python
# Thống kê dữ liệu thiếu trong cột 'DT'
print("Số lượng giá trị thiếu trong cột 'DT' trước khi xử lý:")
print(df['DT'].isnull().sum())

# Hiệu chỉnh dữ liệu thiếu: điền giá trị 0 cho cột 'DT'
df['DT'] = df['DT'].fillna(0)

# Kiểm tra lại sau khi xử lý
print("\nSố lượng giá trị thiếu trong cột 'DT' sau khi xử lý:")
print(df['DT'].isnull().sum())

# Hiển thị vài hàng để xác nhận thay đổi (nếu có giá trị NaN trước đó)
print("\nMột số hàng của cột 'DT' sau khi xử lý thiếu:")
display(df[df['DT'] == 0].head())
```

### 5. Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean.

Tiếp theo, chúng ta sẽ xử lý dữ liệu thiếu cho các biến điểm số còn lại, bao gồm `T1` và các biến điểm khác từ `T1` đến `N6`, cũng như `DH1`, `DH2`, `DH3`, bằng cách điền giá trị trung bình (mean) của từng cột.

```python
# Identify score columns (assuming T1 to N6 are scores and DH1, DH2, DH3 are also scores)
score_columns = [col for col in df.columns if col[0] in ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N'] and col[1:].isdigit()]
score_columns.extend(['DH1', 'DH2', 'DH3'])

print("Columns to check for missing values and impute with mean:", score_columns)

# Check and impute missing values for 'T1' and other score columns with their respective means
for col in score_columns:
    if df[col].isnull().sum() > 0:
        mean_value = df[col].mean()
        df[col] = df[col].fillna(mean_value)
        print(f"Filled missing values in '{col}' with mean: {mean_value:.2f}")
    else:
        print(f"No missing values found in '{col}'.")

# Verify no more missing values in these columns after imputation
print("\nMissing values after imputation:")
for col in score_columns:
    print(f"{col}: {df[col].isnull().sum()}")
```

### 7. Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12.

Chúng ta sẽ tính điểm trung bình môn (TBM) cho các năm lớp 10, 11 và 12 dựa trên công thức đã cho: `TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10`.

```python
# Create TBM1 (Lop 10)
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10

# Create TBM2 (Lop 11)
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10

# Create TBM3 (Lop 12)
df['TBM3'] = (df['T3']*2 + df['L3'] + df['H3'] + df['S3'] + df['V3']*2 + df['X3'] + df['D3'] + df['N3']) / 10

print("New columns TBM1, TBM2, TBM3 created (first 5 rows):")
display(df[['TBM1', 'TBM2', 'TBM3']].head())
```

### 8. Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1, TBM2 và TBM3

Chúng ta sẽ phân loại học lực (Yếu, Trung bình, Khá, Giỏi, Xuất sắc) dựa trên điểm TBM đã tính cho từng năm.

```python
# Function to categorize academic performance
def categorize_performance(score):
    if score < 5.0:
        return 'Y' # Yếu
    elif 5.0 <= score < 6.5:
        return 'TB' # Trung bình
    elif 6.5 <= score < 8.0:
        return 'K' # Khá
    elif 8.0 <= score < 9.0:
        return 'G' # Giỏi
    else: # score >= 9.0
        return 'XS' # Xuất sắc

# Apply the function to create XL1, XL2, XL3
df['XL1'] = df['TBM1'].apply(categorize_performance)
df['XL2'] = df['TBM2'].apply(categorize_performance)
df['XL3'] = df['TBM3'].apply(categorize_performance)

print("New columns XL1, XL2, XL3 created (first 5 rows):")
display(df[['TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head())
```

### 9. Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình sang thang điểm 4 của Mỹ.

Chúng ta sẽ sử dụng phương pháp Min-Max Normalization để chuyển đổi điểm TBM từ thang điểm 10 sang thang điểm 4. Công thức chuyển đổi là `TBM_US = TBM_VN * 0.4`.

```python
# Create US_TBM1, US_TBM2, US_TBM3 using Min-Max Normalization to 4-point scale
# Original range: 0-10, new range: 0-4
# Formula: (TBM - 0) / (10 - 0) * (4 - 0) + 0 = TBM / 10 * 4 = TBM * 0.4
df['US_TBM1'] = df['TBM1'] * 0.4
df['US_TBM2'] = df['TBM2'] * 0.4
df['US_TBM3'] = df['TBM3'] * 0.4

print("New columns US_TBM1, US_TBM2, US_TBM3 created (first 5 rows, 4-point scale):")
display(df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head())
```

### 10. Tạo biến kết quả xét tuyển (KQXT) nhằm xác định sinh viên đậu/rớt.

Chúng ta sẽ xác định sinh viên đậu (giá trị '1') hoặc rớt (giá trị '0') dựa trên điểm `DH1`, `DH2`, `DH3` và khối thi (`KT`) theo các tiêu chí đã định.

```python
# Function to calculate admission status (KQXT)
def calculate_kqxt(row):
    khoi_thi = row['KT']
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']

    score = 0.0
    if khoi_thi in ['A', 'A1']:
        score = (dh1*2 + dh2 + dh3) / 4
    elif khoi_thi == 'B':
        score = (dh1 + dh2*2 + dh3) / 4
    else: # Khối khác
        score = (dh1 + dh2 + dh3) / 3

    return 1 if score >= 5.0 else 0

# Apply the function to create KQXT
df['KQXT'] = df.apply(calculate_kqxt, axis=1)

print("New column KQXT created (first 10 rows):")
display(df[['KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head(10))
```
